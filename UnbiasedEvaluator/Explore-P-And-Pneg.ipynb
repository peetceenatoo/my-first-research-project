## Explore

```python
# Open cml-yahoo-test-pos-unbiased_evaluate_partial.pickle with rb from folder Dataset
with open('./Dataset/cml-yahoo-test-pos-biased_evaluate_partial.pickle', 'rb') as handle:
    P = pickle.load(handle)

# Open cml-yahoo-test-neg-unbiased_evaluate_partial.pickle with rb from folder Dataset
with open('./Dataset/cml-yahoo-test-neg-biased_evaluate_partial.pickle', 'rb') as handle:
    P_neg = pickle.load(handle)

# Get P and P_neg keys

P_keys = list(P.keys())
P_neg_keys = list(P_neg.keys())

# For each key in P, get the corresponding key in P_neg and compare the values

for key in P_keys:
    if key in P_neg_keys:
        print(P[key])
        print(P_neg[key])
        print("\n-------------------\n")
```
